Now we have to do "Text-Vectorization"
    
First Step:    

Now we have converting a text into vector and then try to find a similarities between two vectors.

Then remove the stopwords from the list

in the "tags" feature, There is  number of similar words like- 'big','bigger','biggest'
so the stemming function is convert it into a single word that is "big" only.

So for that we will use NLTK library

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
df['tags']
```

```python
df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()

vectors
```

Now we have to calculate the distance between the vectors in "5000 dimensional space" for that measurement we will use "Cosine Function" instead of "Euclidean Distance"


In high dimensional Euclidean distance is not good measure to be calculation.

At high dimensions, Euclidean distance loses pretty much all meaning.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #pass the all vectors here
similarity
```

Now we have to create one function, according to that function on movie the function will recommend you a 5 similar movies.

Steps Performing:

Step1: first we have to find the index of that given movie.

Step2: Then according to that index the function will cames into the similarity matrix (above mentions function) and took out that perticular index of movie.


Step3: Then according to that index we got the similaraties and we have to sort the distance of similarities.

Step4: After applied sorting function we got "More similar movies" in Beginning and "Less similar movies" in Ending.

And when we do sorting the elements of list are spread randomly, So that the index position will lose

Step5: So for that problem we have to grab the index of movie (Creat the index position according to distance) so for that we will use "Enumerate function"

Step6: According to the steps we will use Enumerate function and then we follow the process of sorting.


Step5: Then the recommend function fetch the "Top 5" movies from that list and recommend to user.

Now you can see the list of touple below i.e. shown the vectordistances of movies 1st, 2nd, 3rd and so on.....

```python
(list(enumerate(similarity[0])))
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])

#So now you can see the similarity in vectors or in movies (In the beginning there are the similar movies)
```

```python
#Now sortout top 5 movies from the list

sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]           #fetched the index
    distances = similarity[movie_index]                        #find the distances of movie index
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)    #Because the system recommend title of the movie not a index of the title 
        
        
    
```

```python
recommend('Avatar')
```

```python
recommend('John Carter')
```

```python
recommend('Batman Begins')
```
